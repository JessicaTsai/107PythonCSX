```python
import requests
```

```python
res = requests.get("https://www.gomaji.com/search?keyword=%E9%A5%97%E9%A3%9F%E5%A4%A9%E5%A0%82")
```

```python
print(res.text)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
print(soup.prettify())
```

```python
soup.select('.relative')
```

```python
for item in soup.select('.relative'):
    print(item.select('.ellipsis')[1].text)
```
